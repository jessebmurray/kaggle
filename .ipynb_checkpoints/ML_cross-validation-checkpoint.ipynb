## Cross-validation

Cross-validation is used to provide a more accurate estimate of the model by expanding the amount of validation data to be the entire dataset. 

This is accomplished by 'folding' the data into, say, five folds. Then, five expriments or tests are run. In each case, one of the folds is set aside as validation data and the other four folds are used as training data. We then obtain five results of how well the model predicted each of the five folds of the validation data. 

Cross-validation should be used for *small datasets*, where the extra computational burden is small, and the susceptibility of the model to a small validation dataset is large.

```python
import pandas as pd

# Read the data
data = pd.read_csv('../data/melb_data.csv')

# Select subset of predictors
cols_to_use = ['Rooms', 'Distance', 'Landsize', 'BuildingArea', 'YearBuilt']
X = data[cols_to_use]

# Select target
y = data.Price
```

We can define a pipeline that imputes missing values and uses a random forest model to make predictions. 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),  # default is to fill with mean
                           ('model', RandomForestRegressor(n_estimators=50, random_state=0))])
```

Explanation of possible scorings:

https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter

Explanation of estimators:

https://scikit-learn.org/stable/developers/develop.html

```python
from sklearn.model_selection import cross_val_score

# Multiply by -1 because sklearn calculates the negative of the MAE
# The estimator (pipeline) is the first argument
scores = -1 * cross_val_score(pipeline, X, y, cv=5, scoring='neg_mean_absolute_error')

print('MAE scores: ', scores, sep='\n')
```

```python
print('Average MAE score (across experiments):')
print(scores.mean())
```

```python
def get_score(n_estimators):
    """Return the average MAE over 3 CV folds of random forest model.
    
    Keyword argument:
    n_estimators -- the number of trees in the forest"""
    
    pipeline = Pipeline([('preprocessor', SimpleImputer()), 
                         ('model', RandomForestRegressor(n_estimators=n_estimators, random_state=0))])
    
    scores = cross_val_score(pipeline, X, y, cv=3, scoring='neg_mean_absolute_error')
    
    return - scores.mean()
```

```python
get_score(10)
```

```python
scores = {val: get_score(val) for val in range(50, 450, 50)}
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')

plt.title('Mean three-fold cross-validation for different numbers of estimators')
plt.xlabel('Number of estimators')
plt.ylabel('Mean absolute error')
plt.plot(list(scores.keys()), list(scores.values()))
plt.show()
```
